**Check for null values in all the columns**

```python
nulls = data.isna().sum()
nulls.sort_values(ascending = False).head(80)
```

It can be seen that some columns have a very high number of null values. Let's calculate the percentage of the null rows compared to the whole number of rows.

```python
#Checking the percentage of null rows:
nulls_percentage = nulls * 100 / len(data)
nulls_percentage.sort_values(ascending = False)
```

```python
nulls_percentage_df = pd.DataFrame(nulls_percentage, columns = ['percentage'])
nulls_percentage_df
```

**Identify columns that over 85% missing values**



```python
high_nulls = list(nulls_percentage_df[nulls_percentage_df['percentage'] > 0.85].index)
high_nulls
```

**Remove those columns from the dataframe**

```python
#putting the columns that we want to keep in a list
data2 = []
for col in data.columns:
    if col not in high_nulls:
        data2.append(col)
data2
```

```python
#removing the columns with high null values
data2_df = data.drop(columns = high_nulls)
```

```python
data2_df
```

**Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest**

```python
data2_df['GENDER'].value_counts()
```

```python
#We will first replace those null values using the code below:
data2_df['GENDER'] = data2_df['GENDER'].replace(' ', 'F')
```

```python
#putting the value counts and categories in a dataframe
data_gender1 = pd.DataFrame(data2_df['GENDER'].value_counts().reset_index().values, columns=["GENDER", "number"])
data_gender1
```

```python
#defining a new dataframe for the categories with the frequency less thatn 1716
data_gender_new = data_gender1[data_gender1['number'] < 1716]
```

```python

data_gender_new_1= list(data_gender_new['GENDER'])
```

```python
data_gender_new_1
```

```python
#defining a function for bucketing
def reduce_categories(x):
    if x in data_gender_new_1:
        return 'other'
    else:
        return x
```

```python
#calling the function
data2_df['GENDER'] = data2_df['GENDER'].apply(reduce_categories)
```
